```python
import cv2
import os
import random
```

```python
input_folder = "/content/drive/MyDrive/input/images"
output_folder = "/content/drive/MyDrive/blurimages/"
```

```python
file_list = os.listdir(input_folder)
```

```python
p = 0.6
blur_limit = (3, 9)
```

```python
i = 0
for filename in file_list:
    # 이미지 파일 경로 생성
    file_path = os.path.join(input_folder, filename)

    # 이미지를 로드합니다.
    image = cv2.imread(file_path)

    # p 값에 따라 Median Blur 적용 여부를 결정합니다.
    if random.random() < p:
        # 랜덤한 커널 크기를 선택합니다.
        ksize = random.choice([3, 5, 7, 9])

        # 중간값 블러를 적용합니다.
        blurred_image = cv2.medianBlur(image, ksize)

        # 결과 이미지를 저장합니다.
        output_file_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_file_path, blurred_image)
        i += 1
        print(f"Completed Median Blur for {filename}, saved to {output_file_path}")
        print(i)
    else:
        # Median Blur를 적용하지 않고 원본 이미지를 저장합니다.
        output_file_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_file_path, image)
        i+=1
        print(f"No Median Blur applied to {filename}, saved to {output_file_path}")
        print(i)
```
